### Analyse tweets from Twitter with hashtags related to healthcare in the UK.

```python
# Libraries and settings needed for analysis
import pandas as pd
import seaborn as sns

# Set figure size.
sns.set(rc={'figure.figsize':(15, 12)})

# Set the plot style as white.
sns.set_style('white')

# Maximum column width to display
pd.options.display.max_colwidth = 200
```

```python
# Load the data set.
tw = pd.read_csv('tweets.csv')

# View the output.
tw
```

```python
# Explore the metadata.
tw.info()
```

```python
# Explore the data set.
# Determine the descriptive statistics of the data set.
tw.describe()
```

```python
# Explore the data set.
# Number of missings values in tw data set. 
tw.isnull().sum()
```

```python
# Would it be useful to only look at retweeted and favourite tweet messages?
'''
YES! A retweet or liked tweet acts as an affirmation. 
It can be assumed that retweets and liked tweets have a shared consensus. 
A shared opinion can be analysed for further insight. 
Moreover, focusing on retweets and liked tweets reduces the possibility of encountering outliers. 
'''
```

```python
# Create a new DataFrame containing only the text.
tw_text = tw[['tweet_full_text']]

# View the output.
tw_text
```

```python
# Create a variable and assign an empty list to it.
tags = []

# Loop through the messages, and create a list of values containing the # symbol.
for y in [x.split(' ') for x in tw_text['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())

# View the ouput.
tags
```

```python
# Display the first 30 records.
tags_30 = tags[0:30]

# View the output. 
tags_30
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
tweets_data = pd.DataFrame(tags, columns = ['tweets'])

# View the output. 
tweets_data
```

```python
# Group "tweets" and rename the columns
tweets_data = tweets_data.groupby(['tweets'])['tweets'].count().reset_index(name="count")

# View the output. 
tweets_data
```

```python
# Check "count" datatype. 
tweets_data.info()
```

```python
# Display records where the count is larger than 10.
tweets_10 = tweets_data[(tweets_data['count']>10)] 

#View the output. 
tweets_10
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
sns.barplot(x = 'tweets',
            y = 'count',
            data = tweets_10)
 
# Show the plot. 
plt.show()
```

```python
# Create copy of tweets_10 DataFrame. 
tweets_10_cleaned = tweets_10.copy()

# View the output. 
tweets_10_cleaned
```

```python
# Create subset of tweets_10 by removing overreprestnted tweets. 
overrep_index = tweets_10["tweets"].isin(['#covid', '#doctor', '#education', '#healthcare', '#hospitals', 
                                          '#innovation', '#job', '#marketing', '#medicine', '#news', 
                                          '#nurses', '#patientcare', '#pharmaceutical', '#science', '#strategy', '#technology', 
                                          '#usa', '#vaccine', '#wellness', '#womeninmedicine'])

overrep_tweets = tweets_10[overrep_index]

# View the output. 
overrep_tweets
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
sns.barplot(x = 'tweets',
            y = 'count',
            data = overrep_tweets)
 
# Show the plot. 
plt.show()
```

```python
# Prepare your workstation.
# Load the appointments_regional.csv file.
ar = pd.read_csv('appointments_regional.csv')

# View the DataFrame.
ar.head(5)
```

```python
# Filter the data set to only look at data from 2021-08 onwards.
ar_date = ar.loc [(ar['appointment_month'] >= '2021-08')]

# View the output. 
ar_date.head()
```

**Question 1:** Should the NHS start looking at increasing staff levels? 

```python
# Create an aggregated data set to review the different features.
ar_agg = ar_date[['appointment_month', 'appointment_status', 'hcp_type', 'appointment_mode', 
                  'time_between_book_and_appointment', 'count_of_appointments']]

# View the output.
ar_agg.head()
```

```python
# Determine the total number of appointments per month.
ar_agg_month = ar_agg.groupby(['appointment_month'])['count_of_appointments'].agg('sum').reset_index()

# View the output.
ar_agg_month
```

```python
# Add a new column to indicate the average utilisation of services.
# Monthly aggregate / 30 to get to a daily value.
ar_average = ar_agg_month.assign(average_utilisation = lambda x: x.count_of_appointments / 30)

# View the output.
ar_average
```

```python
# Convert the appointment_month to string data type for ease of visualisation.
ar_average['appointment_month'].astype(str)
```

```python
# Plot sum of count of monthly visits.
sns.lineplot(x = 'appointment_month', y = 'average_utilisation', data = ar_average, ci = None)
```

**Question 2:** How do the healthcare professional types differ over time?

```python
# Create a lineplot to answer the question. 
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data = ar_agg[ar_agg['hcp_type'].isin(['GP', 'Other Practice staff', 'Unknown' ])],
             hue ='hcp_type', ci=None)
```

**Question 3:** Are there significant changes in whether or not visits are attended?

```python
# Create a line plot to answer the question.
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data = ar_agg[ar_agg['appointment_status'].isin(['Attended'])],
             hue ='appointment_status', ci=None)
```

**Question 4:** Are there changes in terms of appointment type and the busiest months?

```python
# Identify the 3 busiest months. 
busiest_months = ar_agg_month.sort_values(by = 'count_of_appointments', ascending = False)

# View the output. 
busiest_months
```

```python
# Create a lineplot to answer the question based on busiest_months DataFrame findings. 
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data = ar_agg[ar_agg['appointment_mode'].isin(['Face-to-Face', 'Home Visit', 'Telephone', 
                                                            'Unknown', 'Video/Online'])],
             hue ='appointment_mode', ci=None)
```

**Question 5:** Are there any trends in time between booking an appointment?

```python
# Create a line plot to answer the question.
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data = ar_agg[ar_agg['time_between_book_and_appointment'].isin(['Same Day', '1 Day', '2 to 7 Days', 
                                                                             '8  to 14 Days', '15  to 21 Days', 
                                                                             '22  to 28 Days', 'More than 28 Days', 
                                                                             'Unknown / Data Quality'])],
             hue ='time_between_book_and_appointment', ci=None)
```
